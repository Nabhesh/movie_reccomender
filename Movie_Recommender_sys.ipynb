```python
#GEtting names from genre list

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# Converting string into integer for looping
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
#Getting 3 Actors from cast column

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            
            L.append(i['name'])
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
movies['crew'][1]
```

```python
# Fetch Director

def fetch_dir(obj):
    L = []
   
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            
            L.append(i['name'])
            break
        
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_dir)
```

```python
movies.head()
```

```python
movies['overview'][1]
```

```python
#Coverting overview to list

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['genres'] + movies['cast'] + movies['crew'] + movies['keywords']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# Stemming : Finding & replacing similar meaning words
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return "  ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][1]
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# Converting words into vectors and removing stop words

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
# Calculating cosine distance as we have 4800 dimension and it will fail for euclidian distance

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] #fetching index of the movie
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
